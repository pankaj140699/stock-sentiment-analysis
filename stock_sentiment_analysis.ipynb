```python
train=df[df['Date']<='2014-12-31']
test=df[df['Date']>'2014-12-31']
```

```python
data=train.iloc[:,2:27]
```

```python
data.head(5)
```

```python
data=data.replace("[^a-zA-Z]"," ",regex=True)
```

```python
data1=data.replace("b "," ",regex=True)
```

```python
data.head(4)
```

```python
data1.head(2)
```

```python
data1.columns=[str(i) for i in range(1,26)]
```

```python
data1.head(3)
```

```python
for i in data1.columns:
    data1[i]=data1[i].str.lower()
```

```python
data1.head(3)
```

```python
headlines=[]
for i in range(data1.shape[0]):
    headlines.append(' '.join(x for x in data1.iloc[i,0:25]))
```

```python
data1.iloc[0,0]
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement bag of word
```

```python
cv=CountVectorizer(ngram_range=(2,2))
```

```python
traindataset=cv.fit_transform(headlines)
```

```python
# implement random forest
```

```python
rd=RandomForestClassifier(n_estimators=200,criterion='entropy')
rd.fit(traindataset,train['Label'])
```

```python
# prediction
```

```python
test_transform=[]
for i in range(len(test.index)):
    test_transform.append(' '.join(x for x in test.iloc[i,2:27]))
test_dataset=cv.transform(test_transform)
predictions=rd.predict(test_dataset)
```

```python
# check accuracy
```

```python
from sklearn.metrics import confusion_matrix,f1_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
```

```python
print(matrix)
```

```python
score=f1_score(test['Label'],predictions)
```
